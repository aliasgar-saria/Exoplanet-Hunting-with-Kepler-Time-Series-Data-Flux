# Project - Exoplanets Hunting with Kepler labelled time series data


In this Jupyter notebook, we delve into the fascinating realm of exoplanet detection using Kepler-labelled time series data. Our primary objective is to develop a reliable K-Nearest Neighbors (KNN) classification model that can accurately identify stars with exoplanets based on their flux data, which represents the changes in light intensity.


The data describe the change in flux (light intensity) of several thousand stars. Each star has a binary label of 2 or 1. 2 indicated that that the star is confirmed to have at least one exoplanet in orbit; some observations are in fact multi-planet systems.

As you can imagine, planets themselves do not emit light, but the stars that they orbit do. If said star is watched over several months or years, there may be a regular 'dimming' of the flux (the light intensity). This is evidence that there may be an orbiting body around the star; such a star could be considered to be a 'candidate' system. Further study of our candidate system, for example by a satellite that captures light at a different wavelength, could solidify the belief that the candidate can in fact be 'confirmed'.

Data Source: https://www.kaggle.com/datasets/keplersmachines/kepler-labelled-time-series-data

Working with KNN Classification

```python
# Extracting independent (X), dependent (Y)
```

```python
x = train_df.drop(['LABEL'], axis = 1)
y = train_df.LABEL
```

```python
# Splitting data in Train,Test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 4)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

```python
np.min(x_train_sc), np.max(x_train_sc)
```

Choosing the correct value for 'K'

```python
from sklearn.neighbors import KNeighborsClassifier as KNC

# Define a range of K values to try
k_values = range(1, 10)

# Initialize variables to store the best K value and minimum error
best_k = None
min_error = float('inf')

# Initialize a list to store the error rates
error_rates = []

# Iterate over each K value
for k in k_values:
    # Create a KNN classifier with the current K value
    knn = KNC(n_neighbors=k)
    
    # Fit the classifier to the training data
    knn.fit(x_train, y_train)
    
    # Predict the labels for the test data
    y_pred = knn.predict(x_test)
    
    # Calculate the error (misclassification rate)
    error = 1 - accuracy_score(y_test, y_pred)
    
    # Store the error rate for the current K value
    error_rates.append(error)
    
    # Check if the current K value gives a lower error
    if error < min_error:
        min_error = error
        best_k = k

# Print the best K value and its corresponding minimum error
print("Best K value:", best_k)
print("Minimum error:", min_error)

# Plot the error rates
plt.plot(list(k_values), error_rates, marker='o')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
plt.title('Error Rate for Different K Values')
plt.show()
```

```python
# K = 4
knn_classifier = KNC(n_neighbors = 4, metric = 'minkowski', p = 2)

# Fitting Model
knn_classifier.fit(x_train_sc, y_train)


# Predict
y_pred = knn_classifier.predict(x_test_sc)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, auc

print("Validation accuracy of KNN:- ", accuracy_score (y_test, y_pred))
print ()
print("Classification report:- \n", classification_report(y_test, y_pred))

#Confusion matrix 
plt.figure(figsize=(15, 11))
plt.subplots_adjust(wspace = 0.3)
plt.suptitle("KNN Performance before handling the imbalance in the data", color = 'r', weight = 'bold')
plt.subplot (221)
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True, cmap="Set2", fmt = "d", linewidths=3, cbar = False, 
            xticklabels=['nexo','exo'], yticklabels=['nexo', 'exo'], square = True)
plt.xlabel("True Labels", fontsize = 15, weight = 'bold', color = 'tab:pink')
plt.ylabel("Predicited Labels", fontsize = 15, weight = 'bold', color = 'tab:pink')
plt.title("CONFUSION MATRIX", fontsize=20, color = 'm')

#ROC curve and Area under the curve plotting
predicting_probabilites = knn_classifier.predict_proba(x_test_sc)[:,1]
fpr, tpr, thresholds = roc_curve (y_test, predicting_probabilites)
plt.subplot(222)
plt.plot(fpr, tpr, label = ("AUC :", auc(fpr, tpr)),color = "g")
plt.plot([1,0], [1,0], "k--")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC - CURVE & AREA UNDER CURVE", fontsize=20, color = 'm')
plt.show
```

Handling Imbalance in the Data

```python
x,y
```

Balancing Data for correct evaluation metrics

```python
#pip install imblearn
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()
x_ros, y_ros  = ros.fit_resample(x,y)
```

```python
y_ros.value_counts().plot(kind = 'bar')
```

```python
from collections import Counter
print(f"Before ROS:- {Counter(y)} \n After ROS:- {Counter(y_ros)}")
```

```python
# Splitting data in Train,Test - Balanced data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_ros,y_ros,test_size = 0.3, random_state = 4)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

```python
np.min(x_train_sc), np.max(x_train_sc)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNC

# K = 4
knn_classifier = KNC(n_neighbors = 4, metric = 'minkowski', p = 2)

# Fitting Model
knn_classifier.fit(x_train_sc, y_train)


# Predict
y_pred = knn_classifier.predict(x_test_sc)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, auc

print("Validation accuracy of KNN:- ", accuracy_score (y_test, y_pred))
print ()
print("Classification report:- \n", classification_report(y_test, y_pred))

#Confusion matrix 
plt.figure(figsize=(15, 11))
plt.subplots_adjust(wspace = 0.3)
plt.suptitle("KNN Performance after balancing the data", color = 'r', weight = 'bold')
plt.subplot (221)
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True, cmap="Set2", fmt = "d", linewidths=3, cbar = False, 
            xticklabels=['nexo','exo'], yticklabels=['nexo', 'exo'], square = True)
plt.xlabel("True Labels", fontsize = 15, weight = 'bold', color = 'tab:pink')
plt.ylabel("Predicited Labels", fontsize = 15, weight = 'bold', color = 'tab:pink')
plt.title("CONFUSION MATRIX", fontsize=20, color = 'm')

#ROC curve and Area under the curve plotting
predicting_probabilites = knn_classifier.predict_proba(x_test_sc)[:,1]
fpr, tpr, thresholds = roc_curve (y_test, predicting_probabilites)
plt.subplot(222)
plt.plot(fpr, tpr, label = ("AUC :", auc(fpr, tpr)),color = "g")
plt.plot([1,0], [1,0], "k--")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC - CURVE & AREA UNDER CURVE", fontsize=20, color = 'm')
plt.show
```

After addressing the class imbalance issue by balancing the data using the Random OverSampler technique, we observed a significant improvement in the accuracy of the KNN classifier for exoplanet detection.


Prior to balancing the data, the KNN classifier exhibited lower accuracy results when trained on the imbalanced dataset. However, by employing the Random OverSampler to balance the data, the KNN classifier achieved higher accuracy scores, indicating its enhanced ability to classify both positive (exoplanet) and negative (non-exoplanet) samples correctly.


The confusion matrix analysis further supports the effectiveness of data balancing. It reveals that the KNN classifier, trained on the balanced dataset, displayed improved performance in accurately identifying exoplanets and non-exoplanets. This is evident from the higher counts of true positives and true negatives, demonstrating a reduction in misclassifications.


Furthermore, the Receiver Operating Characteristic (ROC) curve and the Area Under the Curve (AUC) metric provide additional evidence of the improved performance after balancing the data. The ROC curve illustrates a better trade-off between true positive rate and false positive rate, indicating improved discrimination between exoplanet and non-exoplanet samples. The larger AUC value confirms the classifier's enhanced ability to distinguish between the two classes.


In summary, balancing the data using the Random OverSampler technique proved highly beneficial in improving the accuracy and performance of the KNN classifier for exoplanet detection. The classifier demonstrated enhanced capabilities in correctly identifying both positive and negative samples, as evidenced by the confusion matrix, ROC curve, and AUC metric analyses.
